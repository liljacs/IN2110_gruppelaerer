```python
# Vi starter med å importere..
# CountVectorizer bruker vi til å gjøre dataene om til en vektorrepresentasjon
# Den gjør dataen om til en matrise som viser frekvensen av hvert token
# Vi importerer scatter_plot fra "helpers"-mappa i v22-repoet

from sklearn.feature_extraction.text import CountVectorizer
from helpers.oblig1 import scatter_plot
```

```python
# så åpner vi filene
# dette er så og si tilfeldige språklige data, laget for å være litt ulike
# vi bruker "with .. as" for å slippe å lukke filene manuelt

# vi tokeniserer dataen først ved hvert linjeskift. 
# 'data' vil være filobjektet

with open("snjo_ut.txt", "r", encoding="utf-8") as data:
    snjo = [x for x in data.read().split("\n") if x != ""]

with open("davvi_ut.txt","r", encoding="utf-8") as data:
    davvi = [x for x in data.read().split("\n") if x != ""]
        
with open("nyheit_ut.txt","r", encoding="utf-8") as data:
    nyheit = [x for x in data.read().split("\n") if x != ""]
```

```python
# Så skiller vi ut dataene og navna på klassene (labels):

data = snjo + davvi + nyheit # denne bruker vi når vi skal transformere datene våre 
labels = ["snø" for x in snjo] + ["samisk" for x in davvi] + ["nyhet" for x in nyheit]
```

```python
# Så initialiserer vi vectorizeren
vectorizer = CountVectorizer()

# Vi bruker fit_transform() til å transformere dataene våre til vektorer
X = vectorizer.fit_transform(data)
```

```python
# Til slutt kan vi lage plottet vårt med labelsene våre og vektorrepresentasjonene
scatter_plot(X, labels)
```

### Hva forteller plottet oss?
Legg merke til hvordan dataene våre er distribuert. 
Vi ser at tekstene med labels "snø" og "nyhet", ligger nærme hverandre i vektorrommet.
Dette er fordi disse dokumentene har mange ord til felles, til forskjell fra den samiske teksten, som jo er på samisk.
